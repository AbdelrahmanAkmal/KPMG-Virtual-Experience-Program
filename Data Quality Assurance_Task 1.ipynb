## Section 2 - Data quality assurance

In This section we will check every dataset for inconvenient datatypes, null entries, duplicated rows. 

```python
def unique_values_object_col(data,columns):
    '''
    This function is to check for duplicates values but with same meaning in the object type columns by showing the unique values of the columns
    ex. United States of America is same as USA or U.S.A. 
    '''
    for col in columns:
        if data[col].dtype == 'O':
            print(f'Exploration column is {col}')
            print(data[col].unique())
            print('------------------------------------------------------------')
```
